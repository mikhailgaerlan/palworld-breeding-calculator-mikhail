```python
import pandas as pd
import os
import networkx
from itertools import combinations_with_replacement
```

```python
breeding_matrix = pd.read_csv(os.path.join('data', 'breeding_matrix.csv'), header=0, skiprows=1, usecols=range(1, 140), index_col=0, nrows=138)
breeding_matrix
```

```python
breeding_matrix.to_pickle(os.path.join('data', 'breeding_matrix.pkl'))
```

```python
breeding_matrix = pd.read_pickle(os.path.join('data', 'breeding_matrix.pkl'))
```

```python
pals = breeding_matrix.index
```

```python
pal_combos = combinations_with_replacement(pals, 2)
pal_combos = pd.MultiIndex.from_tuples(pal_combos, names=['parent1','parent2'])
```

```python
import tqdm
breeding_combos = pd.DataFrame(0, dtype=int, columns=pal_combos, index=pal_combos)
for pal_combo in tqdm.tqdm(pal_combos):
    breeding_combos.loc[pal_combo, breeding_matrix.loc[*pal_combo]] = 1
pd.to_pickle(breeding_combos, os.path.join('data', 'breeding_combos.pkl'))
```
