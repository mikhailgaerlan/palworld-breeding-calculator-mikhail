```python
import pandas as pd
import tqdm
import networkx
```

```python
breeding_matrix = pd.read_csv('breeding_matrix.csv', header=0, skiprows=1, usecols=range(1, 140), index_col=0, nrows=138)
breeding_matrix
```

```python
pals = breeding_matrix.index
```

```python
from itertools import combinations_with_replacement
```

```python
pal_combos = combinations_with_replacement(pals, 2)
pal_combos = pd.MultiIndex.from_tuples(pal_combos, names=['parent1','parent2'])
```

```python
breeding_combos = pd.DataFrame(0, dtype=int, columns=pal_combos, index=pal_combos)
for pal_combo in tqdm.tqdm(pal_combos):
    breeding_combos.loc[pal_combo, breeding_matrix.loc[*pal_combo]] = 1
```

```python
breeding_combos
```

```python
breeding_graph = networkx.from_pandas_adjacency(breeding_combos)
```

```python
class PalworldShortestBreeding:

    def __init__(self, matrix_source, combo_source):
        self.breeding_matrix =  pd.read_csv(matrix_source, header=0, skiprows=1, usecols=range(1, 140), index_col=0, nrows=138)
        self.breeding_combos = pd.read_pickle(combo_source)

        self.pals = self.breeding_matrix.index
        self.pal_combos = combinations_with_replacement(pals, 2)

        self.graph = self.build_graph()
    
    def build_graph(self):
        return networkx.from_pandas_adjacency(self.breeding_combos)
```

```python
def get_combos(pal, exception=None):
    if exception:
        return [pal_combo for pal_combo in pal_combos if pal in pal_combo and exception not in pal_combo]
    return [pal_combo for pal_combo in pal_combos if pal in pal_combo]
```

```python
def get_shortest_path(breeding_graph, source, target):
    try:
        return networkx.shortest_path(breeding_graph, source=source, target=target)
    except networkx.NetworkXNoPath:
        return []
```

```python
def get_subgraph(breeding_graph, pal):
    nodes = [pal_combo for pal_combo in pal_combos if pal not in pal_combo]
    nodes.append((pal, pal))
    return breeding_graph.subgraph(nodes)
```

```python
parent = 'Lamball'
child = 'Anubis'
sources = get_combos(parent, exception=child)

paths = [get_shortest_path(get_subgraph(breeding_graph, child), source, (child, child)) for source in sources]
```

```python
path_lengths = [len(path) for path in paths if len(path) > 0]
min_path_length = min(path_lengths) if len(path_lengths) > 0 else 0
```
